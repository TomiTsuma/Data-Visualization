```python
import matplotlib.pyplot as plt
```

```python
slices = [7,2,3,13]
activities = ['sleeping','working','eating','playing']
cols = ['c','m','r','b']

plt.pie(slices,
       labels = activities,
       colors = cols,
       startangle =45,
       shadow = True,
       explode =(0,0.1,0,0,),
       autopct = '%1.1f%%')

plt.title('Pie Chart')
plt.show()
```
